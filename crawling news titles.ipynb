```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
import re
import requests 

import warnings # 警告模組
warnings.filterwarnings('ignore')
```

```python
#與網頁取得連線

url = 'https://udn.com/news/breaknews/1'

headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)\
             AppleWebKit/537.36 (KHTML, like Gecko)\
             Chrome/91.0.4472.77 Safari/537.36', 
          }


res = requests.get(url, headers = headers)

#print(res.text)
```

```python
#解析HTML程式碼

soup = BeautifulSoup(res.text, 'html.parser')

#print(soup.prettify())
```

```python
#尋找欲擷取資訊的程式碼

items = soup.find_all('section', attrs={'class':'cate-list__subheader'})

#print(len(items), '\n')

#print(items, '\n')

#print(type(items[0]))
```

```python
#在程式碼中擷取資訊

pattern = 'href=\"(/news/breaknews/1/\S+)\">(\w+)'

txt = re.findall(pattern,str(items))

print(txt)
```

```python
#建立欲擷取之主題的list，並擷取每個主題的新聞標題

news = ['精選','要聞','國際','財經','數位']

for k in range(len(txt)):
    
    if txt[k][1] in news:

        url = 'https://udn.com/'

        headers = { 'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)\
                     AppleWebKit/537.36 (KHTML, like Gecko)\
                     Chrome/91.0.4472.77 Safari/537.36', 
                  }


        res = requests.get(url+txt[k][0], headers = headers)

        soup = BeautifulSoup(res.text, 'html.parser')

        #print(res.text)

        items = soup.find_all('section', attrs={'id':'breaknews'})

        #print(len(items))

        #print(items, '\n')

        #print(type(items[0]))

        title = items[0].select('div.story-list__text a')    

        titles = []

        for j in range(0,len(title),2):
            titles.append(title[j]['title'])

        df = pd.DataFrame(titles, columns = [txt[k][1]])

        df.index += 1 

        print(df)   
```
